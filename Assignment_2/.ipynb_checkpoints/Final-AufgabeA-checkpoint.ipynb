### Import und Dataset Laden 

```python
import numpy as np 
import pandas as pd
import matplotlib
import missingno as msno
from fancyimpute import mice

from sklearn.impute import SimpleImputer

# read in the San Francisco building permits data
sfPermits = pd.read_csv("./Building_Permits.csv")

# set seed for reproducibility
np.random.seed(0)
```

```python
sfPermits.sample(5)
```

### Analyse der Fehlerhäufigkeit

```python
# Calculate total number of cells in dataframe
totalCells = np.product(sfPermits.shape)

# Count number of missing values per column
missingCount = sfPermits.isnull().sum()

# Calculate total number of missing values
totalMissing = missingCount.sum()

# Calculate percentage of missing values
print("Es fehlen", round(((totalMissing/totalCells) * 100), 2), "%", " der Werte im Dataset")
```

```python
header = list(sfPermits.columns.values)
missingCount[header]
```

```python
missingPercent = (round(((missingCount[header] / sfPermits.shape[0]) * 100), 2))
print("Fehlende Daten (in Prozent):\n" , missingPercent)
```

```python
%matplotlib inline

missingdata_df = sfPermits.columns[sfPermits.isnull().any()].tolist()
msno.matrix(sfPermits[missingdata_df])
```

### Bereinigen der Daten durch löschen aller Datan mit NaN Inhalt

```python
remNa = sfPermits.dropna()
remNa
```

### Bereinigen der Daten durch löschen der Spalte welche NaN enthält

```python
sfPermitsCleanCols = sfPermits.dropna(axis=1)
sfPermitsCleanCols.head()
```

#### Anzahl der übrig gebliebenen Spalten

```python
print("Gesamtanzahl Spalen: %d \n" % sfPermits.shape[1])
print("Spalten nach löschen von NaN: %d" % sfPermitsCleanCols.shape[1])
```

##### --> keine Lösung da zu viele Spalten verloren gehen

### Bereinigen durch Austausch der NaN Werte mit 0

```python
imputeSfPermits = sfPermits.fillna(method='ffill', axis=0).fillna("0")
print("Spalten nach bereinigen von NaN: %d" % imputeSfPermits.shape[1])
```

### Bereinigen durch manuelles Löschen von Spalten
_Vorgehensweise:_
- Löschen aller Spalten mit mehr als 70% fehlender Werte
- Löschen aller Zeilen mit mehr als 70% fehlender Werte
- Ersetzen durch Andere Werte des Datasets



```python
manDel = sfPermits.drop(["Street Number Suffix","TIDF Compliance","Unit Suffix","Site Permit", "Fire Only Permit","Voluntary Soft-Story Retrofit","Unit"], axis = 1)
manDel = manDel.fillna(method ='bfill')
print("Spalten nach bereinigen von NaN: %d" % manDel.shape[1])
```

### Bereinigen der Daten mit SimpleImputer

```python
imp = SimpleImputer(strategy="most_frequent")
simp= imp.fit_transform(sfPermits)
simp.head()
```
