### Analyse der Fehlerhäufigkeit

```python
# Calculate total number of cells in dataframe
totalCells = np.product(sfPermits.shape)

# Count number of missing values per column
missingCount = sfPermits.isnull().sum()

# Calculate total number of missing values
totalMissing = missingCount.sum()

# Calculate percentage of missing values
print("Es fehlen", round(((totalMissing/totalCells) * 100), 2), "%", " der Werte im Dataset")
```

```python
missingCount[['Street Number Suffix', 'Zipcode']]
```

```python
print("Fehlende Daten (in Prozent) in Spalte: Street Number Suffix =", (round(((missingCount['Street Number Suffix'] / sfPermits.shape[0]) * 100), 2)))
print("Fehlende Daten (in Prozent) in Spalte: Zipcode =", (round(((missingCount['Zipcode'] / sfPermits.shape[0]) * 100), 2)))
```
