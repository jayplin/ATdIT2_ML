### more sophisticated data clean-up
our strategy: before we split some columns into more columns to derive more meaning from them or even do the one-hot-encoding, we delete unnecessary columns first. 
- a)drop the "proposed"-colums


```python
#looking at the dataset you'll see that there are several columns appaering twice: 
#once as the "proposed" version and once as the "existing" version. 
# let's compare them: 
print((sfPermits['Existing Use'] == sfPermits['Proposed Use']).sum() / len(sfPermits['Proposed Use']))
# and since this was a String; let's do this for "Existing Units" (float used like
#an Integer... ), too: 
print((sfPermits['Existing Units'] == sfPermits['Proposed Units']).sum() / len(sfPermits['Proposed Units']))
# check for the others: 
print((sfPermits['Revised Cost'] == sfPermits['Estimated Cost']).sum() / len(sfPermits['Estimated Cost']))
print((sfPermits['Number of Existing Stories'] == sfPermits['Number of Proposed Stories']).sum() / len(sfPermits['Number of Proposed Stories']))

print((sfPermits['Proposed Construction Type'] == sfPermits['Existing Construction Type']).sum() / len(sfPermits['Proposed Construction Type']))
```

result: as "existing..." and "proposed..." seem to overlap at least 70% ore more, we'll drop the "proposed" columns-except for the costs, which seem to differ quite often :D

### Inverse Data Clean-Up: split date columns
now, that we reduced columns (because the'd either contained too many NaNs, were in the wrong context or because there had been similar columns), in a next step, we'll split date fields. A single date is difficult to correlate to current status or so. However, a single year/month/day (maybe) might be useful. 


```python
dates = [ 'Permit Creation Date',
    'Current Status Date',
    'Filed Date',
    'Issued Date',
    'First Construction Document Date',
    'Permit Expiration Date']

def split_dates_into_three(datename):
    name = datename[0:-4]
    sfPermits[name+'Year'] = sfPermits[datename].astype(str).str[-4:]
    sfPermits[name+'Year'] = pd.to_numeric(sfPermits[name+'Year'], errors='coerce')
    sfPermits[name+'Day'] = sfPermits[datename].astype(str).str[3:5]# US date format: month, day, year
    sfPermits[name+'Day'] = pd.to_numeric(sfPermits[name+'Day'], errors='coerce')
    sfPermits[name+'Month'] = sfPermits[datename].astype(str).str[0:2]
    sfPermits[name+'Month'] = pd.to_numeric(sfPermits[name+'Month'], errors='coerce')
    
for i in dates:
    split_dates_into_three(i)

sfPermits.drop(dates, axis=1, inplace=True)#since now we have the dates in the split version already. 
sfPermits.columns
```

### Imputations
we should make clear if a value is an "original" value or if it has been imputed. 


```python
imputed_permits = sfPermits.copy()
cols_w_missing_values = (col for col in sfPermits.columns #columns with missing values
                                 if sfPermits[col].isnull().any())
for col in cols_w_missing_values:
    imputed_permits[col + '_was_missing'] = imputed_permits[col].isnull()
    
#show percentage of missing values per column: (did this in exploration before, but meanwhile our columns changed quite a bit...)

totalCells = np.product(imputed_permits.shape)
# Count number of missing values per column
missingCount = imputed_permits.isnull().sum()

# Calculate total number of missing values
totalMissing = missingCount.sum()

header = list(imputed_permits.columns.values)

missingPercent = (round(((missingCount[header] / imputed_permits.shape[0]) * 100), 2))
print("Missing data in & :\n" , missingPercent)
```

```python
print(sfPermits.shape)
print(imputed_permits.shape)
```

since we introduced several new columns before starting with the imputation, we got 19 new columns. 

#### imputation for Strings/categorical data (use the mode): 
although of different data types, they are treated equally, since it's not target-aimed to calculate the mean of categorical data - be it numerical or not....

```python
columns_for_mode = ['Existing Use','Plansets','Existing Construction Type','Supervisor District',
                   'Neighborhoods - Analysis Boundaries','Zipcode' ]

for column in columns_for_mode:
    imputed_permits[column].fillna(imputed_permits[column].mode()[0], inplace=True)
    
```

#### imputation for Strings/categorical data (use the mode): 
basically, the different date fields, cost (2x), nor of stories and existing units are numerical data: 


```python
for column in imputed_permits.columns:
    if ( imputed_permits[column].isnull().sum() == 0 ):
        continue
imputed_permits[column].fillna( imputed_permits[column].mean(skipna=True) , inplace=True)
```

```python
missingCount = imputed_permits.isnull().sum()
# Calculate total number of missing values
totalMissing = missingCount.sum()
missingPercent = (round(((missingCount[header] / imputed_permits.shape[0]) * 100), 2))
print("Missing data in & :\n" , missingPercent)
```

for whatever reason, this didn't work for all  columns that were numerical, so let's try it ion another way: 

```python
numerical_columns = ['Number of Existing Stories','Estimated Cost','Revised Cost',
    'Existing Units','Issued Year','Issued Day','Issued Month','First Construction Document Year', 
       'First Construction Document Day', 'First Construction Document Month', 'Permit Expiration Year',
         'Permit Expiration Day', 'Permit Expiration Month' 
                    ]
for col in numerical_columns:
    imputed_permits[col].fillna( imputed_permits[col].mean(skipna=True) , inplace=True)
```
