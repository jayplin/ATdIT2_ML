## Imports

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
```

## Download and prepare data

```python
test_size = 1000
```

```python
train = tf.keras.datasets.mnist.load_data()[0]
X_train, Y_train = train[0], train[1]

test = tf.keras.datasets.mnist.load_data()[1]
X_test, Y_test = test[0], test[1]
X_test = X_test[:test_size]
Y_test = Y_test[:test_size]

n_train = X_train.shape[0]
n_test = X_test.shape[0]

m = X_train.shape[1]**2
```

```python
X_train = X_train.reshape([n_train, m])
X_test = X_test.reshape([n_test, m])
print(X_test.shape)
print(Y_test.shape)
```

## Do KMeans Algorithm 

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics.cluster import homogeneity_score
def doKmeans(iter):
  n_digits = len(np.unique(Y_test))

  kmeans = KMeans(init='k-means++', n_clusters=n_digits, n_init=10, n_jobs=-1, verbose=0, max_iter=iter)

  X_test_local = X_test.reshape([n_test, m])
  #print(X_test_local.shape)
  #print(Y_test.shape)

  kmeans.fit(X_test)

  for y in range(10):
      print("\nCluster: %d" % (y))
      cluster = (np.array([i for i, x in enumerate(kmeans.labels_) if x == y]))
      labels_cluster_1 = Y_test[cluster]
      len_cluster = len(labels_cluster_1)

      unique, counts = np.unique(labels_cluster_1, return_counts=True)
      counts_dict = dict(zip(unique, counts))
      majority_class = max(counts_dict.items(), key=lambda k: k[1])[0]

      print(counts_dict)
      print("Percentage of majority class: %f" % (max(counts)/float(len_cluster)))
      print("Majority class is: %d" % (majority_class))


      print("Laenge: ", len_cluster)
      print("Homogeneity Score for %d: %f" %( iter, homogeneity_score(Y_test, kmeans.labels_)))
      
      
```

- Hinweis Cluster 0: 9 und 7 haben die gleiche Anzahl. Es gibt also keine Majority Class für Cluster 0.
- Es gibt keine Cluster für die Zahlen 5 und 9
